# USING TREND API

The Trend methods allows you to list the current trends for any geographical location with their respective WOEID code

```python
def woeid_details(id_no):
    sfo_trends_india = api.trends_place(id=id_no)
    try:
        names=[]
        urls=[]
        promoted_contents=[]
        querys=[]
        tweet_volumes=[]
        for i in range(0,49):
            names.append(sfo_trends_india[0]['trends'][i]['name'])
            urls.append(sfo_trends_india[0]['trends'][i]['url'])
            promoted_contents.append(str(sfo_trends_india[0]['trends'][i]['promoted_content']))
            querys.append(str(sfo_trends_india[0]['trends'][i]['query']))
            tweet_volumes.append(str(sfo_trends_india[0]['trends'][i]['tweet_volume']))
    except:
        print(f'Query no %s exceeded'%(i))
    Codedet=pd.DataFrame(zip(names,urls,querys,tweet_volumes,promoted_contents),columns=['Names','URL','QUERY','TWEET VOLUME','PROMOTED CONTENT'])
    return Codedet
    
```

# STREAM API

Streaming allows you to actively watch for tweets that match certain criteria in real time. This means that when there aren‚Äôt any new tweet matching the criteria, then the program will wait until a new tweet is created and then process it.

To use streaming you have to create two objects:

The stream object uses the Twitter API to get tweets that match some criteria. This object is the source of tweets that are then processed by a stream listener.
The stream listener receives tweets from the stream.

```python
import json
import tweepy
import sys
from datetime import datetime,date
class MyStreamListener(tweepy.StreamListener):
    def __init__(self, api):
        self.api = api
        self.me = api.me()

    def on_status(self, tweet):
            print(f"{tweet.created_at.strftime('%H:%M:%S')}:{tweet.user.name}:\t{tweet.text}:\n")
            
        
try:
    tweets_listener = MyStreamListener(api)
    stream = tweepy.Stream(api.auth, tweets_listener)
    stream.filter(track=["staraikelungal"])
except KeyboardInterrupt:
    print("\t STREAMING ENDED \t")
    
```
